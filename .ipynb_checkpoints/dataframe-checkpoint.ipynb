```python
df = pd.read_csv("object_detection_results.csv")
```

```python
grouped = df.groupby("Frame")
```

```python
for frame_number, group in grouped:
    print(f"Frame {frame_number}:")
    print(group)
```

```python
import pandas as pd
import numpy as np

# Assuming you already have your DataFrame 'df' and it's grouped by 'Frame'

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=["Frame Group", "Object Name", "Average X1", "Average Y1", "Average X2", "Average Y2"])

# Define the size of each frame group
group_size = 10

# Iterate through the groups
for group_num, (frame_number, group) in enumerate(grouped):
    if group_num % group_size == 0:
        # Create a DataFrame to store the results for the current group
        group_result = pd.DataFrame(columns=["Frame Group", "Object Name", "Average X1", "Average Y1", "Average X2", "Average Y2"])
        frame_group = frame_number // group_size

    # Calculate the average for each object within the group
    avg_values = group.groupby("Object Name")[["X1", "Y1", "X2", "Y2"]].mean().reset_index()
    avg_values["Frame Group"] = frame_group

    # Append the results to the group DataFrame
    group_result = group_result.append(avg_values, ignore_index=True)

    if (group_num + 1) % group_size == 0:
        # If this is the last frame in the current group, add the group results to the main results DataFrame
        result_df = result_df.append(group_result, ignore_index=True)

# If an object is not present in all frames within a group, replace the NaN with Null
result_df = result_df.fillna("Null")

print(result_df)
```
