```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("object_detection_results-v8.csv")

# Define a function to update the coordinates of each object
def update_coordinates(group):
    # Calculate the updated coordinates for each object
    x1 = group['X1'].min()
    y1 = group['Y1'].min()
    x2 = group['X2'].max()
    y2 = group['Y2'].max()
    
    # Create a new row with the updated coordinates
    updated_row = group.iloc[0].copy()
    updated_row['X1'] = x1
    updated_row['Y1'] = y1
    updated_row['X2'] = x2
    updated_row['Y2'] = y2
    
    return updated_row

# Group the DataFrame by every 10 frames and apply the update_coordinates function
result = df.groupby(df.index // 10).apply(update_coordinates)

# Reset the index of the resulting DataFrame
result = result.reset_index(drop=True)

# Now, 'result' contains the combined frames with updated coordinates for each object
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("object_detection_results-v8.csv")

# Drop duplicates based on both "Frame" and "Object Name" columns and keep the first occurrence
df = df.drop_duplicates(subset=['Frame', 'Object Name'], keep='first')

# Group the DataFrame by every 10 frames
grouped = df.groupby(df.index // 10)

# Define a function to compute the average of X1, Y1, X2, Y2 for repeated objects
def compute_average(group):
    return group.groupby('Object Name')[['X1', 'Y1', 'X2', 'Y2']].mean()

# Apply the compute_average function to each group and reset the index
result = grouped.apply(compute_average).reset_index()

# Reset the index of the resulting DataFrame
result = result.reset_index(drop=True)

# Now, 'result' contains 40 rows with the average coordinates for each object
```
